# Notebook to convert videos from .mov to .mp4

```python
import os, cv2
```

## Configuration

```python
input_file = '/Users/cyrilmonette/Desktop/EPFL 2018-2026/PhD - Mobots/Media/Videos/PhanomenaFilm3.mov'
```

## Main code

```python
# Open the video file
cap = cv2.VideoCapture(input_file)
# Get the width, height, and frames per second (fps) of the video
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
# Define the codec and create a VideoWriter object to save the video as MP4
fourcc = cv2.VideoWriter_fourcc(*'avc1')
output_file = os.path.splitext(input_file)[0] + '.mp4'
# Throw an error if the output file already exists
if os.path.exists(output_file):
    raise FileExistsError(f'Output file already exists: {output_file}')
out = cv2.VideoWriter(output_file, fourcc, fps, (width, height))
# Read each frame from the video and write it to the output file
while True:
    ret, frame = cap.read()
    if not ret:
        break
    out.write(frame)
# Release the video objects
cap.release()
out.release()
print(f'Video converted and saved as: {output_file}')
```
